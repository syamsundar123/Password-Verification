```python
from tkinter import *
win=Tk()
win.title("Calculator")
win.config(bg="pink")

e=Entry(win,width=35,borderwidth=3)
e.grid(row=0,column=1,columnspan=3,padx=10,pady=10)

def Button_click(number):
    current=e.get()
    e.delete(0,END)
    e.insert(0,str(current)+str(number))
def Clear():
    e.delete(0,END)



#Math Functions
def add():
    first_number=e.get()
    global F_N
    global math
    math="addition"
    F_N=float(first_number)
    e.delete(0,END)
    
    
def sub():
    first_number=e.get()
    global F_N
    global math
    math="subtraction"
    F_N=float(first_number)
    e.delete(0,END)
    
def mul():
    first_number=e.get()
    global F_N
    global math
    math="multiplication"
    F_N=float(first_number)
    e.delete(0,END)
def div():
    first_number=e.get()
    global F_N
    global math
    math="division"
    F_N=float(first_number)
    e.delete(0,END)
    
def Equal():
    second_number=e.get()
    e.delete(0,END)
    if math=="addition":
        e.insert(0,F_N+float(second_number))
    if math=="subtraction":
        e.insert(0,F_N-float(second_number))
    if math=="multiplication":
        e.insert(0,F_N*float(second_number))
    if math=="division":
        e.insert(0,(F_N)/(float(second_number)))
        
#defining Buttons
    
btn1=Button(win,text="1",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(1))
btn2=Button(win,text="2",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(2))
btn3=Button(win,text="3",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(3))
btn4=Button(win,text="4",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(4))
btn5=Button(win,text="5",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(5))
btn6=Button(win,text="6",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(6))
btn7=Button(win,text="7",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(7))
btn8=Button(win,text="8",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(8))
btn9=Button(win,text="9",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(9))
btn0=Button(win,text="0",padx=40,pady=20,bg="skyblue",command=lambda:Button_click(0))



btn_dot=Button(win,text=".",padx=40,pady=20,bg="skyblue",command=lambda:Button_click("."))
btn_add=Button(win,text="+",font="Helvetica,20,bold",padx=34,pady=20,bg="skyblue",command=add)
btn_sub=Button(win,text="-",font="Helvetica,20,bold",padx=35,pady=20,bg="skyblue",command=sub)
btn_mul=Button(win,text="*",font="Helvetica,20,bold",padx=36,pady=20,bg="skyblue",command=mul)
btn_div=Button(win,text="/",font="Helvetica,20,bold",padx=35,pady=20,bg="skyblue",command=div)
btn_eql=Button(win,text="=",font="Helvetica,20,bold",padx=35,pady=14,bg="skyblue",command=Equal)
btn_clr=Button(win,text="Clear",font="Helvetica,20,bold",padx=63,pady=20,bg="brown",command=Clear)


#Placing Buttons on Screen
btn1.grid(row="3",column="1")
btn2.grid(row="3",column="2")
btn3.grid(row="3",column="3")

btn4.grid(row="2",column="1")
btn5.grid(row="2",column="2")
btn6.grid(row="2",column="3")

btn7.grid(row="1",column="1")
btn8.grid(row="1",column="2")
btn9.grid(row="1",column="3")

btn0.grid(row="4",column="1")
btn_dot.grid(row="4",column="2")
btn_eql.grid(row="4",column="3")

btn_add.grid(row="5",column="1")
btn_sub.grid(row="5",column="2")
btn_mul.grid(row="5",column="3")

btn_div.grid(row="6",column="1")
btn_clr.grid(row="6",column="2",columnspan="2")
#building mainwindow
win.mainloop()
```
